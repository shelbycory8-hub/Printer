```python
import os
import subprocess
from IPython.display import Image, display

# 1. Install OpenSCAD and Xvfb (Virtual Framebuffer)
print("Installing dependencies for high-end rendering...")
os.system("sudo apt-get update -qq")
os.system("sudo apt-get install -y openscad xvfb -qq > /dev/null")
print("Installation complete. Generating master assembly visuals...")

# 2. OpenSCAD Code for the 7-Part Master Assembly
scad_code = """
$fn = 60;

// Helper module to generate 3D text labels
module manual_label(part_name) {
    color("Black")
    translate([0, -120, -20])
    linear_extrude(height=2)
    text(part_name, size=14, font="Arial:style=Bold", halign="center");
}

// 1. CoreXY Frame & Gantry
module part_corexy_frame() {
    color("DimGray") {
        // Top horizontal rail
        translate([-50, 0, 50])  cube([150, 10, 10], center=true);
        // Bottom horizontal rail
        translate([0, 0, -50])   cube([150, 10, 10], center=true);
        // Left vertical post
        translate([-70, 0, 0])   cube([10, 10, 120], center=true);
        // Right vertical post
        translate([0, -75, 0])   cube([10, 10, 120], center=true);
    }
    // Pulley cylinders at top corners
    color("Silver") {
        translate([-40, 65, 20])  cylinder(h=40, r=15, center=true);
        translate([40, 65, 20])   cylinder(h=40, r=15, center=true);
    }
    manual_label("1. CoreXY Gantry");
}

// 2. Progressive Cavity Pump
module part_fluid_pump() {
    // Motor body
    color("Silver")       translate([0, 0, 0])    cylinder(h=45, r=21, center=true);
    // Pump housing
    color("DarkSlateGray") translate([0, 0, -35])  cylinder(h=25, r=21, center=true);
    // Rotor shaft
    color("DodgerBlue")   translate([0, 0, -85])   cylinder(h=75, r=12, center=true);
    // Inlet port
    color("Gold")         translate([15, 0, -60])  rotate([90, 0, 0]) cylinder(h=15, r=4, center=true);
    // Outlet port
    color("Gold")         translate([0, 0, -130])  cylinder(h=15, r=4, center=true);
    manual_label("2. PC Pump");
}

// 3. Extrusion Toolhead & UV Curing
module part_printhead() {
    // Carriage block
    color("Silver")  translate([0, 0, 10])  cube([40, 40, 30], center=true);
    // Four nozzles at 90-degree intervals
    for(i = [0:90:270]) {
        rotate([0, 0, i]) {
            translate([15, 0, -20]) {
                color("DarkGray") cylinder(h=20, r=6, center=true);
                color("Gold")     translate([0, 0, -15]) cylinder(h=10, r1=0.5, r2=2, center=true);
            }
        }
    }
    // UV ring
    color("Indigo") translate([0, 0, -10]) {
        difference() {
            cylinder(h=8, r=30, center=true);
            cylinder(h=10, r=22, center=true);
        }
    }
    manual_label("3. UV Printhead");
}

// 4. Smart Substrate Mandrel
module part_mandrel() {
    // Egg-shaped mandrel body
    color("NavajoWhite") translate([0, 0, 0])   scale([1, 1.25, 1.3]) sphere(r=45);
    // Spindle shaft
    color("Silver")      translate([0, 0, -65])  cylinder(h=15, r=25, center=true);
    // Mounting plate
    color("DarkGray")    translate([0, 0, -80])  cube([40, 40, 15], center=true);
    manual_label("4. Mandrel");
}

// 5. Reservoir System
module part_reservoirs() {
    // Main manifold block
    color("Silver")           translate([0, 0, -50])  cube([120, 40, 80], center=true);
    // Reservoir A (blue — bioink)
    color("LightBlue", 0.7)   translate([-40, 0, 0])  cylinder(h=90, r=15, center=true);
    color("Gold")             translate([-40, 0, 50]) cylinder(h=10, r=5, center=true);
    // Reservoir B (green — support)
    color("LightGreen", 0.7)  translate([0, 0, 0])    cylinder(h=90, r=15, center=true);
    color("Gold")             translate([0, 0, 50])   cylinder(h=10, r=5, center=true);
    // Reservoir C (white — crosslinker)
    color("White", 0.7)       translate([40, 0, 0])   cylinder(h=90, r=15, center=true);
    color("Gold")             translate([40, 0, 50])  cylinder(h=10, r=5, center=true);
    manual_label("5. Reservoirs");
}

// 6. Electronics & Power Supply
module part_electronics() {
    // PSU enclosure
    color("Silver")     translate([0, 30, -20])  cube([120, 60, 40], center=true);
    // Control board
    color("DarkGreen")  translate([0, -30, 0])   cube([100, 40, 20], center=true);
    // Connectors row
    for(x = [-30 : 12 : 30]) {
        color("Black") translate([x, -40, 5]) cube([10, 10, 5], center=true);
    }
    manual_label("6. Electronics");
}

// 7. VOC Extraction System
module part_voc_filter() {
    // Housing
    color("DimGray")  translate([0, 0, 0])   cube([80, 80, 60], center=true);
    // HEPA layer
    color("Black")    translate([0, 0, -30]) cube([50, 50, 10], center=true);
    // Activated carbon layer
    color("White")    translate([0, 0, 10])  cube([40, 40, 10], center=true);
    manual_label("7. VOC Filter");
}

// --- Master Layout: 7 modules spread along X-axis ---
translate([-750, 0, 0])  part_corexy_frame();
translate([-500, 0, 0])  part_fluid_pump();
translate([-250, 0, 0])  part_printhead();
translate([  0,  0, 0])  part_mandrel();
translate([ 250, 0, 0])  part_reservoirs();
translate([ 500, 0, 0])  part_electronics();
translate([ 750, 0, 0])  part_voc_filter();
"""

# 3. Save the OpenSCAD code to a file
scad_file = "master_assembly.scad"
with open(scad_file, "w") as f:
    f.write(scad_code)

# 4. Render using the Xvfb Virtual Display Server
image_file = "master_assembly_layout.png"

render_cmd = [
    "xvfb-run", "-a", "openscad",
    "--camera=0,0,0,55,0,25,1400",   # camera: translate, rotate, distance
    "--imgsize=2560,720",             # wide panoramic canvas for all 7 parts
    "--colorscheme=Tomorrow Night",
    "-o", image_file,
    scad_file
]

print("Rendering high-res master assembly layout...")
subprocess.run(render_cmd, check=True)

# 5. Display the rendered image inline
print("Rendering complete! Here is your 7-part master visual:")
display(Image(image_file))
```
